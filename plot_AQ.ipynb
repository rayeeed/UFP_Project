```python
def PNC_cat(dt):

  lst=[]
  for i in range(dt.shape[0]):

    if dt.PNC[i] <= 25000:
      lst.append('within 25000 (#/cm^3)')
    elif dt.PNC[i] <= 37500:
      lst.append('between 25000-37500 (#/cm^3)') 
    elif dt.PNC[i] <= 50000:
      lst.append('between 37500-50000 (#/cm^3)')
    elif dt.PNC[i] <= 80000:
      lst.append('between 50000-80000 (#/cm^3)')
    else:
      lst.append('more than 80000 (#/cm^3)') 

  Df['PNC_category'] = lst
```
